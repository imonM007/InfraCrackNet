```python
import os
from datasets import Dataset, Features, ClassLabel, Image
import pandas as pd

# Directories where your images and labels are stored
image_dir = "/content/drive/MyDrive/Dataset/image"
label_dir = "/content/drive/MyDrive/Dataset/label"

# Define a function to load images and labels
def load_data(image_dir, label_dir):
    image_files = sorted(os.listdir(image_dir))  # Assuming images and labels have matching filenames
    label_files = sorted(os.listdir(label_dir))

    data = []
    for image_file, label_file in zip(image_files, label_files):
        if image_file.endswith('.jpg') and label_file.endswith('.jpg'):  # Check for jpg format
            image_path = os.path.join(image_dir, image_file)
            label_path = os.path.join(label_dir, label_file)

            data.append({'image': image_path, 'label': label_path})

    return pd.DataFrame(data)
```

```python
# Load data into a pandas DataFrame
df = load_data(image_dir, label_dir)

# Define features for the dataset
features = Features({
    'image': Image(),
    'label': Image()
})

# Create Hugging Face dataset
dataset = Dataset.from_pandas(df, features=features)
```

```python
dataset.push_to_hub("Imon007/CrackNet")
```
